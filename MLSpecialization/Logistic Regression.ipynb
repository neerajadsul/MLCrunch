# Classification using Logistic Regression
* binary classification (two classes or categories): 
    - positive/negative, pass/fail
        - positive: email is spam, tumour is malignant
        - negative: email is not spam, tumour is benign
    - yes/no, True/False, 1/0
    
## Logistic Regression

### Sigmoid function (logistic function)
$g(z) = \frac{1}{1+e^{-z}}$ when $0 < g(z) < 1$

$f_{W,b} = W \cdot X + b$

Therfore using logistic regression as activation function,

$ f_{W,b} = g(W \cdot X + b) = g(z)$

where $z = W \cdot X + b$

Expanding,

$ f_{W,b} = 1/[1 + e^{-(W \cdot X + b)}] $ indicates the probablity that class is 1.


In research papers/books it may be written as: 

$ f_{W,b} = (y=1 |X; w,b) $ indicates the probablity that class is 1.


## Decision Boundary

Let's classify 

* as 1 when $g(z) \geq 0.5 $, 
* otherwise 0 when $g(z) < 0.5$.

This implies, 

* Class 1 when $z \geq 0$ and 
* Class 0 when $z<0$

**Note**: The threshold value does not need to be at the middle of the range. It could be 0.2 or 0.8 based on the problem.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.e, 2*np.e, 100)
y = lambda z,a,b: a/(b + np.exp(-z))
fig, ax = plt.subplots()
ax.plot(x, y(x, 1 , 1), label="z,1,1")
ax.plot(x, y(x, 1.2 , 1), label="z,1.2,1")
ax.plot(x, y(x, 1 , 2), label="z,1,2")
ax.plot(x, y(x, 1.2 , 2), label="z,1.2,2")
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
plt.grid(True)
ax.legend()
ax.set_title("Sigmoid Function")
```
